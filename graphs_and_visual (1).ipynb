```python
# highest win margin by runs
df.loc[df['result']=='runs','result_margin'].max()

# lowest win margin by runs
df.loc[df['result']=='runs','result_margin'].min()
```

```python
# Find the maximum run-margin value
max_margin_runs = df.loc[df['result'] == 'runs', 'result_margin'].max()

# Get the full match details for that margin
match_details = df.loc[(df['result'] == 'runs') & (df['result_margin'] == max_margin_runs)]

match_details
```

```python
min_margin_runs=df.loc[df['result']=='runs','result_margin'].min()

match_details=df.loc[(df['result']=='runs') & (df['result_margin']==min_margin_runs)]

match_details
```
